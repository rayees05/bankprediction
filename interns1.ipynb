```python
import pandas as pd
df=pd.read_csv('E://Video/mushrooms.csv',header=0)
# df1=pd.read_csv('E://Video/hour.csv')
print(df.head())
```

```python
df=df.dropna()
```

```python
df=pd.get_dummies(df,drop_first=True)
print(df)
```

```python
df.isnull().sum()
```

```python
x,y=df.drop(labels=['class_p'],axis=1),df['class_p']
```

```python
from sklearn.model_selection import train_test_split
from lazypredict.Supervised  import LazyClassifier
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=12,test_size=0.2)
clf=LazyClassifier(predictions=True)
models,predictions=clf.fit(xtrain,xtest,ytrain,ytest)
pd.DataFrame(models)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model=AdaBoostClassifier()
model.fit(xtrain,ytrain)
```

```python
pred=model.predict(xtest)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
pd.DataFrame({'actual':ytest,'predictions':pred})
```
