```python
import pandas as pd
credits=pd.read_csv('E:/tmdb_5000_credits.csv')
movies=pd.read_csv('E:/tmdb_5000_movies.csv')
credits.head()
```

```python
movies.head()
```

```python
credits_renamed=credits.rename(index=str,columns={'movie_id':'id'})
movies_merge=movies.merge(credits_renamed,on='id')
movies_merge.head()
```

```python
movies_merge.drop(columns=['homepage','production_companies','original_language','title_x','title_y'])
```

```python
R=movies_merge['vote_average']
v=movies_merge['vote_count']
m=movies_merge['vote_count'].quantile(0.70)
C=movies_merge['vote_average'].mean()

```

```python
movies_merge['weghted_avg']=((R*v)+(C*m))/(v+m)
movies_merge.head(20)
```

```python
movies_merge_df=movies_merge.sort_values('weghted_avg',ascending=False)
movies_merge_df[['original_title','vote_average','vote_count','weghted_avg','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
w=movies_merge_df.sort_values('weghted_avg',ascending=False)
axi=sns.barplot(x=w['weghted_avg'].head(10),y=w['original_title'].head(10))
plt.title("best movies by avg vote")
plt.xlabel('weighed average')
plt.ylabel('orginal title')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
```

```python
cols=['user_id','item_id','rating','timestamp']
df=pd.read_csv('u.data',sep='\t',names=cols)
movies=pd.read_csv('C:/Users/user/OneDrive/Desktop/Movie_Id_Titles')
df=pd.merge(df,movies, on='item_id')
df.head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
ratings['num_ratings']=df.groupby('title')['rating'].count()
ratings.head()
```

```python
moviemat=df.pivot_table(index='user_id',columns='title',values='rating').fillna(0)
moviemat.head()
```

```python
ratings.sort_values('num_ratings',ascending=False).head()
```

```python
ratings[ratings['num_ratings']>100].count()
```

```python
similer_movie=moviemat['Fargo (1996)']
pred_movie=moviemat.corrwith(similer_movie)
pred_movie
```

```python
def mov_name(name):
    data=moviemat[name]
    corr_mov=pd.DataFrame(moviemat.corrwith(data),columns=(['correlation']))
    corr_mov.dropna(inplace=True)
    corr_mov=corr_mov.join(ratings['num_ratings'])
    corr_mov=corr_mov[corr_mov['num_ratings']>100].sort_values('correlation',ascending=False)
    corr_mov=corr_mov.iloc[1:,2:]
    corr_mov=corr_mov.head(5)
    return corr_mov
```

```python
mov_name('Young Guns II (1990)')
```

```python
from scipy.sparse import csr_matrix
moviemat_matrix=csr_matrix(moviemat.values)

from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric='cosine',algorithm='brute')
model.fit(moviemat_matrix)
```

```python
moviemat_matrix.shape
```

```python
index=np.random.choice(moviemat.shape[0])
print(index)
D,I=model.kneighbors(moviemat.iloc[index,:].values.reshape(1,-1),n_neighbors=5)
```

```python
print(format(moviemat.index[index]))
```

```python
for i in range (0,len(D.flatten())):
    if i==0:
        print('recomented for {0}:\n'.format(moviemat.index[index]))
    else:
        print('{0}:{1}, with distence{2}:'.format(i,moviemat.index[I.flatten()[i]],D.flatten()[i]))
        
```

```python
net_ser=pd.read_csv('E:/netflix_titles.csv')
net_ser.head(10)
```

```python
net_ser=net_ser.drop(columns=['director','title','type','date_added','listed_in'],axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv=TfidfVectorizer(min_df=3,analyzer='word',max_features=None,token_pattern=r'\w{1,}',ngram_range=(1,3),stop_words='english')
net_ser['description']=net_ser['description'].fillna(' ')
```

```python
tfv_net_ser=tfv.fit_transform(net_ser['description'])
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig=sigmoid_kernel(tfv_net_ser,tfv_net_ser)
```

```python
sig[0]
```

```python
indices=pd.Series(net_ser.index, index=net_ser['show_id']).drop_duplicates()
```

```python
indices
```

```python
def give_rec(showid,sig=sig):
    idx=indices[showid]
    rec_mov=list(enumerate(sig[idx]))
    rec_mov=sorted(rec_mov,key=lambda x:x[1],reverse=True)
    rec_mov=rec_mov[1:11]
    rec_indice=[i[0] for i in rec_mov]
    return net_ser['show_id'].iloc[rec_indice]
```

```python
import pandas as pd
anime=pd.read_csv('E:/anime.csv')
rating_ani=pd.read_csv('E:/rating.csv')
anime.head()
```

```python
rating_ani
```

```python
counts=anime['rating'].value_counts()
anime=anime[anime['rating'].isin(counts[counts>=100].index)]
```

```python
anime_merge=anime.merge(rating_ani,on='anime_id')
anime_merge=anime_merge.drop(columns=['type'],axis=1)
anime_merge.head()
```

```python
com_animi=(anime_merge.groupby(by=['name'])['rating_x'].count().reset_index().rename(columns={'rating_x':'ratingcount'})
           [['name','ratingcount']])
```

```python
com_animi
```

```python
com_animi_df=com_animi.merge(anime_merge, left_on='name',right_on='name',how='left')
com_animi_df
```

```python
popular=50
popul_ani=com_animi_df.query('ratingcount>=@popular')
popul_ani.head()
```

```python
from scipy.sparse import csr_matrix
anine_count=popul_ani.drop_duplicates(['name','user_id'])
anine_count_pv=anine_count.pivot(index='name',columns='user_id',values='rating_x').fillna(0)
anine_count_mt=csr_matrix(anine_count_pv.values)
# anine_count_pv.head()
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric='cosine',algorithm='brute')
model.fit(anine_count_mt)
```

```python
import numpy as np
query_index=np.random.choice(anine_count_mt.shape[0])
# print(query_index)
d,I=model.kneighbors(anine_count_pv.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)
```

```python
for i in range(len(d.flatten())):
    if i==0:
        print('recom for {0}'.format(anine_count_pv.index[query_index]))
    else:
        print('{0}:{1}, with distence{2}'.format(i,anine_count_pv.index[I.flatten()[i],d.flatten()[i]]))
```
